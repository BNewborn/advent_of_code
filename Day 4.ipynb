```python
def passport_string_valid(pp_string):
    pp_list = re.split(r'[ ]|\n',pp_string)
    pp_list = [x for x in pp_list if len(x)>1]
    pp_dict = dict()
#     print(f"pp list {pp_list}")
    for key_val in pp_list:
        key = key_val.split(":")[0]
        val = key_val.split(":")[1]
        pp_dict[key]=val
        
    if len(pp_dict)<=6:
#         print("6 or fewer item")
        return False
    elif len(pp_dict)==7 & ~('cid' in pp_dict.keys()):
#         print("7 items, but cid is missing value")
        return True
    elif len(pp_dict)==8:
#         print("all 8 items")
        return True
    else: 
#         print("missing 1 needed value")
        return False
    
    
```

```python
assert(passport_string_valid('''ecl:gry pid:860033327 eyr:2020 hcl:#fffffd
byr:1937 iyr:2017 cid:147 hgt:183cm''')==True)
assert(passport_string_valid('''iyr:2013 ecl:amb cid:350 eyr:2023 pid:028048884
hcl:#cfa07d byr:1929''')==False)
assert(passport_string_valid('''hcl:#ae17e1 iyr:2013
eyr:2024
ecl:brn pid:760753108 byr:1931
hgt:179cm''')==True)
assert(passport_string_valid('''hcl:#cfa07d eyr:2025 pid:166559648
iyr:2011 ecl:brn hgt:59in''')==False)
```

```python
passport_results = list()
for pp in each_passports:
#     print(pp)
#     print()
    try:
        passport_results.append(passport_string_valid(pp))
    except:
        print(f"Error: \n\t{pp}")
        break
    
print(np.array(passport_results).sum(),"valid passports")
```


    byr (Birth Year) - four digits; at least 1920 and at most 2002.
    iyr (Issue Year) - four digits; at least 2010 and at most 2020.
    eyr (Expiration Year) - four digits; at least 2020 and at most 2030.
    hgt (Height) - a number followed by either cm or in:
        If cm, the number must be at least 150 and at most 193.
        If in, the number must be at least 59 and at most 76.
    hcl (Hair Color) - a # followed by exactly six characters 0-9 or a-f.
    ecl (Eye Color) - exactly one of: amb blu brn gry grn hzl oth.
    pid (Passport ID) - a nine-digit number, including leading zeroes.
    cid (Country ID) - ignored, missing or not.

```python
def passport_string_valid_pt_2(pp_string):
    pp_list = re.split(r'[ ]|\n',pp_string)
    pp_list = [x for x in pp_list if len(x)>1]
    pp_dict = dict()
    
    birthyear_flag = False
    issueyear_flag= False
    expyear_flag= False
    height_flag= False
    haircolor_flag= False
    eyecolor_flag= False
    passport_flag= False
            
            
#     print(f"pp list {pp_list}")
    for key_val in pp_list:
        key = key_val.split(":")[0]
        val = key_val.split(":")[1]
        pp_dict[key]=val
        
        if key=='byr':
            try:
                birthyear_flag = 1920<=int(val)<=2002
            except:
                birthyear_flag = False
                
        elif key=='iyr':
            try:
                issueyear_flag = 2010<=int(val)<=2020
            except:
                issueyear_flag = False
                
        elif key=='eyr':
            try:
                expyear_flag = 2020<=int(val)<=2030
            except:
                expyear_flag = False
                
        elif key=='hgt':
            if val[-2:]=='cm':
                try:
                    height_flag = 150<=int(val[:-2])<=193
                except:
                    height_flag = False
            elif val[-2:]=='in':
                try:
                    height_flag = 50<=int(val[:-2])<=76
                except:
                    height_flag = False
            else:
                height_flag = False
            
        
        elif key=='hcl':
            if len(val)==7:
                try:
                    if re.match(r'\#[0-9a-f]{6}',val):
                        haircolor_flag=True
                except:
                    haircolor_flag = False
            else:
                haircolor_flag = False
                
        elif key=='ecl':
            if val in ['amb','blu','brn','gry','grn','hzl','oth']:
                eyecolor_flag = True
            else:
                eyecolor_flag = False
                
        elif key=='pid':
            if (re.match(r'[0-9]{9}',val)) and (len(val)==9):
                    passport_flag=True
            else:
                passport_flag=False
                
        
    flags = [birthyear_flag #0
             ,issueyear_flag
             ,expyear_flag
             ,height_flag
            ,haircolor_flag
             ,eyecolor_flag
             ,passport_flag]
    print(f"Birthyear: {birthyear_flag}")
    print(f"Issueyear: {issueyear_flag}")
    print(f"Exp Year: {expyear_flag}")
    print(f"Height: {height_flag}")
    print(f"Hair Color: {haircolor_flag}")
    print(f"Eye Color: {eyecolor_flag}")
    print(f"Passport: {passport_flag}")
    
    if sum(flags)==7:
        return True
    else:
        return False
    
    
```
