```python
with open('./data/day6_input.txt', 'r',newline='\n') as reader:
    answers = reader.readlines()
grouped_answers = " ".join(answers).split(" \n ")
grouped_answers[0:4]
```

```python
uniq_answers_by_group = list()
for a in grouped_answers:
    temp_group_list = list()
    for b in a:
        if b not in  [' ','\n']:
            temp_group_list.append(b)
    uniq_answers_by_group.append(len(set(temp_group_list)))
#     print("**"*25)
sum(uniq_answers_by_group)
```
